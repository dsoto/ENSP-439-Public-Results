# Data Collected

The graphs below represent the data files we received from the NetAtmo device that is placed in the ETC. The NetAtmo has been collecting baseline data from the ETC for particular months over the last year for data sets like the ETC's temperature, humidity, CO2 amount, pressure inside the building, electricity gained from solar, total electricity used by the entire building, and the amount of natural gas the ETC uses monthly. 

Our ultimate goal is to accurately predict the efficiency of the solar panels on the roof of the ETC. While some of these graphs may not directly correlate to the solar efficiency, it provides more insight as to how the building is acting and how the electricity from our solar panels is contributing to the electrical output.

# Introduction

- In this study, we try to determine the efficiency of the Environmental Technology Center based on how much energy the building requires to keep the building at a comfortable temperature. We are assuming a comfortable temperature ranges from 70-74 degrees Fahrenheit. 

The Environmental Technology Center was built in _ with energy efficiency in mind. Although the building primarily uses passive solar design to heat and cool the building, it has an additional heating system that requires natural gas. The building has 2,200 square feet of concrete floor with hot water tubing underneath that is heated by natural gas. The heated floor is a major source of energy consumption for the building, using roughly 3500 C/FT of natural gas each month. Since the building is passive solar, the building is able to use occupancy levels and passive solar design to be able to heat the building naturally. However, the floors add an additional temperature to the building, making the temperature frequently uncomfortable.

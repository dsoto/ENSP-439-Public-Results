```python
import pandas as pd
data1 = pd.read_csv('Plug_Load_Logger7916.csv', skiprows=1)
data1
```

```python
temp = data1
temp.describe()
```

```python
# Calculation for average power wihtout any exclusions


 
filtered_power = data1['Power W']
filtered_power = filtered_power[filtered_power != 0]
filtered_power.mean()

#data["Power W"].mean()
```

```python
# This is the average power calculation for when the compressor is on and we have excluded all power values less than 130 and those equal to Zero.
# This cell when run will produce the average power when the compressor is running as well as the number of one minute cycles in this mode.


filtered_power_Compressor = data1['Power W']
filtered_power_Compressor = filtered_power_Compressor[filtered_power_Compressor != 0]
filtered_power_Compressor = filtered_power_Compressor[filtered_power_Compressor > 130] 
filtered_power_Compressor.mean() , filtered_power_Compressor.shape
 
```

```python
filtered_power_Compressor.describe()
```

```python
filtered_power_Compressor.sum() /60 /1000
```

```python
#This Code prints the KWh of cumulative power for compression mode.
PowerW_when_operating = 159.9
data_points_when_operating = 9515

Energy_Total_Compressor = PowerW_when_operating * data_points_when_operating /60 /1000
print('The cummulative energy used while operating the compressor throughout this test sample is equal to', Energy_Total_Compressor, 'KWh')

#this is the same as the line of code beneath, but with different labels for the variables. Answer given is in Kilowatt-hours
```

```python
# This is the average power calculation for when the compressor is off and in idle mode. We have excluded all power values above 70 and equal to 0.
#This cell will produce both the average power in idle mode as well as the number of cycles it was in this mode.


filtered_power_Idle = data1['Power W']
filtered_power_Idle = filtered_power_Idle[filtered_power_Idle != 0]
filtered_power_Idle = filtered_power_Idle[filtered_power_Idle < 70] 
filtered_power_Idle.mean() , filtered_power_Idle.shape
```

```python
#This code calculates the cummulative power when in IDLE. answer is in Kilowwatt

Power_when_Idle = 10.233084566367
data_points_when_Idle = 9259
Energy_Total_Idle = Power_when_Idle * data_points_when_Idle / 60/1000
Energy_Total_Idle

print('The cummulative energy used while operating in Idle mode throughout this test sample is equal to', Energy_Total_Idle, 'KWh')
```

```python
#Instructions: Change variable 'SpecifiedDay' to the last day's value you want to see
#              Change 'currentDate' to the first day's value you want to see (e.g. 20160101 for Jan 1st)
#              Change 'houseNum' to the house number you want to analyze (e.g. 111 for house 1, 112 for house 2)
# Remember: If a day says nan, check it by hand. If program crashes, find the broken day and hand calculate it
#           then start the program again from the day right after
# For program maintenece, send cookies to Kdarb
import pandas as pd
data = pd.read_csv('aggregated_PV_data.csv', parse_dates=True, index_col=0)
import math
houseNum = '152'
specifiedDay = 30
tempTotal = []
currentDate = '20160301'
for i in range (0,specifiedDay): #loops from first day to the specified day
    if math.isnan(data.loc[currentDate,houseNum][len(data.loc[currentDate,houseNum])-1]): #checks if the last value is NaN
        #make it look at previous value
        index = 1
        while math.isnan(data.loc[currentDate,houseNum][len(data.loc[currentDate,houseNum])-1-index]): #goes until the value before the last checked value is a number
            if index <= len(data.loc[currentDate,houseNum])-2:
                index += 1
            else:
                print (currentDate + ' is a blackout day! Please move parameters around this day :)')
                break
        tempTotal.append(data.loc[currentDate,houseNum][len(data.loc[currentDate,houseNum])-1-index]-data.loc[currentDate,houseNum][0]) #adds the closest real number to the end array
    else:
        tempTotal.append(data.loc[currentDate,houseNum][len(data.loc[currentDate,houseNum])-1]-data.loc[currentDate,houseNum][0])
        #This gives you the sum for the desired day
    #increase date by 1 here
    currentDateInt = int(currentDate) #converts the string date into a number so I can work with it
    if currentDateInt == 20160131:
        currentDateInt = 20160201
    elif currentDateInt == 20160229:
        currentDateInt = 20160301
    elif currentDateInt == 20160331:
        currentDateInt = 20160401
    elif currentDateInt == 20160430:
        currentDateInt = 20160501
    elif currentDateInt == 20160531:
        currentDateInt = 20160601
    elif currentDateInt == 20160630:
        currentDateInt = 20160701
    elif currentDateInt == 20160731:
        currentDateInt = 20160801
    elif currentDateInt == 20160831:
        currentDateInt = 20160901
    elif currentDateInt == 20160930:
        currentDateInt = 20161001
    elif currentDateInt == 20161031:
        currentDateInt = 20161101
    elif currentDateInt == 20161130:
        break #breaks out of loop if the last day is reached
    else:
        currentDateInt += 1
        
    currentDate = str(currentDateInt) #converts the integer value for date back into a string
        
roundedTempTotal = [ '%.3f' % elem for elem in tempTotal ] #rounds final array to make it easier to read    
print (roundedTempTotal)
```
